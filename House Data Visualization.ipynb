```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data_Path = os.path.join("AU Housing Data", "AU Real Estate Data.csv")
House_List = pd.read_csv(Data_Path)
House_List
```

```python
Cities = ['Sydney', 'Melbourne', 'Brisbane']
Average_Prices = House_List[House_List['City'].isin(Cities)].groupby('City')['Price'].mean()

plt.bar(Cities, Average_Prices)
sns.set_theme(style='whitegrid')
plt.title('Average Price Comparison between Sydney, Melbourne, and Brisbane')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.show()
```

```python
House_List.boxplot(column='Price', by='City', grid=False, vert=False, patch_artist=True,
             boxprops=dict(facecolor='lightblue'))

plt.title('Price Distribution Across Major Cities')
plt.suptitle('') 
plt.xlabel('Price (AUD)')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='City', y='Price', data=House_List, palette='pastel', hue='City', legend=False)
plt.title('Price Distribution Comparison Among Cities')
plt.xlabel('City')
plt.ylabel('Price (AUD)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SqFt', y='Price', data=House_List, hue='City', palette='Set2', alpha=0.7)
plt.title('Relationship Between Price and Square Footage')
plt.xlabel('Square Footage')
plt.ylabel('Price (AUD)')
plt.tight_layout()
plt.show()
```

```python
Region_Counts = House_List['City'].value_counts()

plt.pie(Region_Counts, labels=Region_Counts.index, autopct='%1.1f%%')
plt.title('Distribution of Real Estate Properties between Regions')
plt.show()
```

```python
plt.hist(House_List['Price'], bins=20, edgecolor='black')
plt.title('Distribution of Real Estate Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
corr_matrix = House_List[['Price', 'SqFt', 'Bedrooms', 'Bathrooms']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Factors Affecting Price')
plt.show()
```

```python
sns.pairplot(House_List[['Price', 'SqFt', 'Bedrooms', 'Bathrooms']])
plt.title('Pair Plot for Price, Property Area, Bedrooms, and Bathrooms')
plt.show()
```

```python
sns.kdeplot(House_List['Price'], fill=True)
plt.title('Price Density Plot')
plt.xlabel('Price')
plt.show()
```

```python
Funnel_Data = House_List['City'].value_counts().sort_values(ascending=False)

plt.bar(Funnel_Data.index, Funnel_Data)
plt.title('Funnel Plot: Real Estate Distribution Channel')
plt.show()
```

```python
sns.regplot(x='SqFt', y='Price', data=House_List)
plt.title('Regression Plot: Price vs Property Area')
plt.show()
```
